# Exploring_the_Data
  Cleaning data

```python
# merging datasets
movies = movies.merge(credits, on = 'title')
```

```python
# sorting nescessary colummns
# -->genres,id,keywords,title,overview,cast,crew

movies = movies[['movie_id','title','overview','genres','keywords','cast','crew']]
```

```python
movies.head()
```

```python
movies.isnull().sum()
```

```python
movies.dropna(inplace=True)
```

```python
movies.duplicated().sum()
```

```python
movies.iloc[0].genres
```

```python
movies.drop_duplicates(inplace=True)
```

```python
# cleaning the columns genres and keywords

import ast     # coverts the str into a proper list
def convert(obj):
    L=[]
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L
```

```python
movies['genres'] = movies['genres'].apply(convert)
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
```

```python
movies.head()
```

```python
# cleaning the columns cast

def convertC(obj):
    L=[]
    counter = 0
    for i in ast.literal_eval(obj):
        if counter != 3:
            L.append(i['name'])
            counter =+ 1
        else:
            break
    return L
```

```python
movies['cast'] = movies['cast'].apply(convertC)
```

```python
# cleaning the column crew

def convertCrew(obj):
    L=[]
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
            L.append(i['name'])
            break
    return L
```

```python
movies['crew'] = movies['crew'].apply(convertCrew)
```

```python
movies.head()
```

cleaned the four columns and now converting the overview string column into a list and then adding the cast and crew onto the overview and again getting back overview into string resulting in a paragraph

```python
movies['overview'] = movies['overview'].astype(str).apply(lambda x: x.split())
```

```python
movies['genres'] = movies['genres'].apply(lambda x: [i.replace(" ","") for i in x])
movies['keywords'] = movies['keywords'].apply(lambda x: [i.replace(" ","") for i in x])
movies['cast'] = movies['cast'].apply(lambda x: [i.replace(" ","") for i in x])
movies['crew'] = movies['crew'].apply(lambda x: [i.replace(" ","") for i in x])
```

```python
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
```

```python
movies.drop(columns = ['overview', 'genres', 'keywords', 'cast', 'crew'], inplace = True)
```

```python
movies.head()
```

```python
# converting the list of tags again into a string
movies['tags'] = movies['tags'].apply(lambda x: " ".join(x))
```

```python
movies.head()
```

```python
movies['tags'] = movies['tags'].apply(lambda x: x.lower())
```

```python
# stemming the tags
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    y = []
    for i in text.split():
            y.append(ps.stem(i))
    string = " ".join(y)
    return string
```

```python
movies['tags'] = movies['tags'].apply(stem)
```

```python
movies['tags'][3]
```

```python
# TEXT VECTORIZATION
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000, stop_words='english')
```

```python
vector = cv.fit_transform(movies['tags']).toarray()
```

```python
vector.shape
```

```python
cv.get_feature_names_out().tolist()
```

```python
from sklearn.metrics.pairwise import cosine_similarity 
```

```python
similarity = cosine_similarity(vector)
```

```python
# similarity[index]
# isthe movies name using index and similarity returns the distances

# sorted(list(enumerate(similarity[7])),reverse=True,key =lambda x: x[1])[1:6] 
#                 enumerate is going to hold the index 
```

```python
sorted(list(enumerate(similarity[0])),reverse=True,key =lambda x: x[1])[1:6] 
```

```python
def recommend(movie):
    movie_index = movies[movies['title'] == movie].index[0]
    distances = similarity[movie_index]
    movies_list = sorted(list(enumerate(distances)),reverse=True,key =lambda x: x[1])[1:6]
    
    for i in movies_list:
        print(movies.iloc[i[0]].title)
    
```

```python
recommend('Batman')
```
